Example 1: Assignment is Not Copying

```python
a = 10
b = a
a = 20
print(b)
```

1. The object is immutable
2. Python rebound the value
3. No other references were changed

Example 2: Stack Frame Behavior

```python
def f(x):
  x = x+1
  print(x)

n = 5
f(n)
print(n)
```

1. The object is immutable
2. The function called did not reassign the ouput to n only showed the result of the function
3. No references were changed

Example 3: Immutable Object Behavior

```python
x = 5
print(id(x))
x+=1
print(id(x))
```

1. The object was immutable
2. Python reassigned the x to 6 and the 5 was probably garbage collected
3. Yes the change affected other references

Example 4: Mutable Object Behavior

```python
lst = [1,2,]
print(id(lst))
lst.append(3)
print(id(lst))
```

1. The object was mutable
2. The object was mutated to add 3
3. No references were changed, IDs were the same

Example 5: Argument Passing (Immutable)

```python
def change(x):
  x = x + 10

a = 5
change(a)
print(a)
```

1. The object was immutable
2. Python rebound the name for local function call
3. References were not changed.

Example 6: Argument Passing (Mutable)

```python
def change(lst):
  lst.append(99)

data = [1,2]
change(data)
print(data)
```

1. The object was Mutable
2. Python mutated "data" to add 99
3. The references were not changed

Example 7: Rebinding vs. Mutation

```python
def reset_wrong(lst):
  lst = []

def reset_right(lst):
  lst.clear()

data = [1,2,3]
reset_wrong(data)
print(data)
reset_right(data)
print(data)
```

1. The object was mutable
2. reset_wrong rebound the object while reset_right mutated it
3. The references only changed when reset_right was used

Example 8: Identity Check

```python
def demo(x):
  print(id(x))
  x.append(5)
  print(id(x))

lst = [1,2]
print(id(lst))
demo(lst)
```

1. The object is mutable
2. Python mutated the object, the reference ID is the same and the value is changed
3. The references were not changed

Example 9: Equality vs Identity

```python
a = [1,2]
b = [1,2]

print(a==b)
print(a is b)
```

1. The object is mutable as its a list
2. Not applicable as == and is are just checking equality and identity
3. Not Applicable

Example 10: Rebinding a Name

```python
x = [1,2]
print(id(x))
x = [3,4]

print(x)
print(id(x))
```

1. The object is mutable
2. Python rebound x to [3,4]
3. The references were changed
